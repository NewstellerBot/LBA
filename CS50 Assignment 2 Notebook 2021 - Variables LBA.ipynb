```python
def partition(A, lo, hi):
    '''
    This function creates the partitions within the array
    and returns the indice of the pivot (i.e., all elements
    after the pivot are greater than the pivot, and all elements
    smaller or equal than the pivot are before)
    '''
    #chosing the pivot
    pivot = A[hi]
    # setting the index of the smaller element
    i = lo - 1
    # iterating over the array
    for j in range(lo, hi):
        # if the element is smaller or equal than the pivot
        if A[j] <= pivot:
            # increment the index of the smaller element
            i += 1
            # swap the elements
            A[i], A[j] = A[j], A[i]
    # swap the pivot with the smaller element
    A[i + 1], A[hi] = A[hi], A[i + 1]
    # return the indice of the pivot
    return i + 1

def quick_sort(A, lo, hi):
    '''
    This function sorts given array, using quick sort algorithm. 
    This algorithm is divide and conquer algorithm. More about the 
    algorithm can be read here:
    https://en.wikipedia.org/wiki/Quicksort#Algorithm
    Here Lomuto partition scheme will be implemented
    '''
    if lo < hi:
        # partitioning the array
        p = partition(A, lo, hi)
        # sorting the left part of the array
        quick_sort(A, lo, p - 1)
        # sorting the right part of the array
        quick_sort(A, p + 1, hi)
    

```

```python
def mean(array):
    '''
    This function as an input takes an array
    and calculates the mean of the values from the array
    '''
    # sorting the array
    quick_sort(array, 0, len(array) - 1)
    # this variable will store the sum of all values from the array
    sum_of_all_values = 0
    # this variable will store the length of the array
    N = 0
    
    for value in array:
        # looping through all the values in the array
        # adding all the values to the sum_of_all_variables
        sum_of_all_values += value
        # adding one to the length of the array for one run of the loop
        N += 1

    # returning the mean
    return sum_of_all_values / N
 
def median(array):
    '''
    This function as an input takes an array
    and calculates the media of the values from the array
    '''
    # sorting the array
    quick_sort(array, 0, len(array) - 1)
    # N will store the length of the array
    # len() returns the length of the given list, tuple, string, etc
    N = len(array)

    # this if statement checks whether the number of elements is divisible by two
    if N % 2 == 0:
        # if it is, it returns the median
        return array[int(N / 2)]
    else:
        # if it isn't it calculates the mean of two middle values
        return (array[(N // 2) + 1] + array[N // 2]) / 2

def mode(array):
    '''
    This function as an input takes an array
    and finds the mode of the values from the array
    '''
    # sorting the array
    quick_sort(array, 0, len(array) - 1)
    # this variable will store the number of occurences of each value
    frequncy = {}
    # looping through all the values in the array
    for value in array:
        # if the value is in the dictionary
        if value in frequncy:
            # adding one to the value
            frequncy[value] += 1
        else:
            # if the value is not in the dictionary set its frequency to 1
            frequncy[value] = 1
    # max() returns the largest item in an iterable
    # key parameter is function that is used to extract a comparison key from each list element
    return max(frequncy, key=frequncy.get)

def stat_range(array):
    '''
    This function as an input takes a sorted array
    and calculates the range of the values from the array
    '''
    # max() returns the largest item in an iterable
    # min() returns the smallest item in an iterable
    # range is the difference between the max and min
    return max(array) - min(array)

def variance(array):
    '''
    This function as an input takes an array
    and calculates the variance of the values from the array
    '''
    # sorting the array
    quick_sort(array)
    # mean of the array
    mean = mean(array)
    # length of the array
    N = len(array)
    # sum of the squares of the differences between the mean and each value
    sum_of_squares_differences = 0
    # looping through all the values in the array
    for value in array:
        # adding the square of the difference between the mean and the value
        sum_of_squares_differences += (value - mean) ** 2
    # returning the variance
    return sum_of_squares_differences / N

def standard_deviation(array):
    '''
    This function as an input takes an array
    and calculates the standard deviation of the values from the array
    '''
    # standard deviation is the square root of the variance
    return variance(array) ** 0.5
```

### 4.1 Matching Napkins Simulation

What is the probability that you are unlucky, and sadly NONE of the napkins are matched to the correct measurement (you guessed all of them wrong)? **Write a simulation in Python to estimate this probability**. Ensure that your code has sufficiently detailed comments to explain the key steps in the solution. **Note**: *Use your own data for this simulation*. You should be able to explain how this function works, including the parameters passed, and why it is relevant for the problem. Read the [Numpy documentation](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to learn more.
- If you have $N$ measurements, you will want to check whether `rand_napkins[i] == napkins[i]`, for each value of `i` from $0$ to $N-1$.

```python
def simulate_matching_napkins(napkins):
    '''
    np.random.choice randomly choses from the given set given number of elements
    from the given iterable. Size parameter determines the number of elements being chosen.
    Replace parameter determines whether they are drawn with replacements.
    The probability distribution is uniform, i.e. each element has equal probability of being chosen.
    '''
    random_napkins = np.random.choice(napkins, size=len(napkins), replace=False)
    # Returns a list with the True and False values if the napkins are matched or not matched respectively
    return sum([1 if napkin == random_napkin else 0 for napkin, random_napkin in zip(napkins, random_napkins)])

def simulate_many_matchings(N, napkins):
    '''
    This function takes an integer N and list of napkins 
    as an input and simulates N matchings of napkins
    '''
    # Creating an empty list to store the results
    results = []
    # Looping through N times
    for i in range(N):
        # Simulating a single matching
        results.append(simulate_matching_napkins(napkins))
    # Returning the results
    return results
```

```python
napkins = [1, 2, 3] # Each napkin has distinct value--there are no duplicates
N = 10000 # Simulate 10000 matchings
results = simulate_many_matchings(N, napkins)
```

```python
'''
I can use only 3 bins here because if 2 napkinss are matched,
then the third one, naturally, must also be matched. Therefore,
there is bin for 0, 1, and 3 matches. 
'''
plt.hist(results, bins=3)    
plt.xlabel('The number of matched napkins')
plt.ylabel('Number of times a napkin was matched')
plt.show()
```

```python
def probability_of_matching(results, n):
    '''
    This function takes a list of results and an integer n
    as an input and calculates the probability of n matches
    '''
    # sum() returns the sum of all the elements in the list
    return sum([1 if result == n else 0 for result in results]) / len(results)
```

```python
num_of_all_matched_napkins = probability_of_matching(results, len(napkins))
print(f'The probability of matching all three napkins is {num_of_all_matched_napkins / N}')
```
